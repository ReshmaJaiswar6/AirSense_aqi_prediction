#common section

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import re
from collections import defaultdict
```

```python
def img_summarizer(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)

    # Collect PNG files
    all_files = [f for f in os.listdir(input_folder) if f.endswith('.png')]

    if not all_files:
      print("No PNG files found.")
      exit()

    # Regex: 02JAN2023
    date_pattern = r"\d{2}[A-Z]{3}\d{4}"

    # Group files by date
    date_groups = defaultdict(list)

    for f in all_files:
      match = re.search(date_pattern, f)
      if match:
          date_groups[match.group(0)].append(f)

    # Process each date separately
    for date, files in date_groups.items():
      running_sum = None
      count = 0

      for f in files:
          img = cv2.imread(os.path.join(input_folder, f), cv2.IMREAD_GRAYSCALE)
          if img is None:
              continue

          if running_sum is None:
              running_sum = np.zeros_like(img, dtype=np.float64)

          running_sum += img / 255.0
          count += 1

      if count == 0:
          continue

      final_avg = running_sum / count

      # Clean plot (NO axes, labels, ticks, colorbar)
      plt.figure(figsize=(6, 6))
      plt.imshow(final_avg, cmap='gray', vmin=0, vmax=1)
      plt.axis('off')

      output_path = os.path.join(output_folder, f"{date}.png")
      plt.savefig(output_path, dpi=300, bbox_inches='tight', pad_inches=0)
      plt.close()

      print(f"{date}: summarized {count} images → saved")
```
