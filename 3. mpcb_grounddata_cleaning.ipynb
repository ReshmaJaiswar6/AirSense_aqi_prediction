#cleaning mpcb 2024 data

```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_excel("/content/drive/MyDrive/Amajor/img: inp- code- op/cpcb_bkc/AQI_daily_2024_Bandra_Kurla_Complex_Mumbai_MPCB_2024.xlsx")
```

```python
month_map = {
    "January": "JAN",
    "February": "FEB",
    "March": "MAR",
    "April": "APR",
    "May": "MAY",
    "June": "JUN",
    "July": "JUL",
    "August": "AUG",
    "September": "SEP",
    "October": "OCT",
    "November": "NOV",
    "December": "DEC"
}

df = df.rename(columns=month_map)
```

```python
df_long = df.melt(
    id_vars=["Day"],
    value_vars=list(month_map.values()),
    var_name="Month",
    value_name="AQI"
)
df2= df_long
df2 = df2[["Day", "Month", "AQI"]]
```

```python
#df2["Day"] = df2["Day"].astype(int)
df2["Month"] = df2["Month"].astype(str)
df2["AQI"] = df2["AQI"].astype(float)

print(df2.head())
```

```python
print(df2.to_string())
```

```python
df2['Day'] = pd.to_numeric(df2['Day'], errors='coerce')
df2 = df2.dropna(subset=['Day'])
df2['Day'] = df2['Day'].astype(int)
```

```python
print(df2.to_string())
```

```python
df2.isna().sum()
```

```python
'''drop illegal dates fr 2024,
29, 30 feb,
31 apr, jun, sept, nov
'''
```

```python
df2.drop(
    df2[
        (df2['Month'] == 'FEB') &
        (df2['Day'].isin([29, 30]))
    ].index,
    inplace=True
)
df2.drop(
    df2[
        (df2['Month'].isin(['APR', 'JUN', 'SEP', 'NOV'])) &
        (df2['Day'] == 31)
    ].index,
    inplace=True
)
```

```python
df2.isna().sum()
```

```python
df2.ffill(inplace=True)
```

```python
df2.isna().sum()
```

```python
df2.bfill(inplace=True)
```

```python
df2.isna().sum()
```

```python
output_path = '/content/drive/MyDrive/Amajor/img_summ/mcpb2024BKC.csv'
df2.to_csv(output_path, index=False)
print(f"DataFrame saved to {output_path}")
```

```python
def prep_values(inp_data, month):
  inp_data = inp_data.apply(lambda row: row.ffill(), axis=1)
  inp_data = inp_data.apply(lambda row: row.bfill(), axis=1)
  op_data= pd.DataFrame(columns= ['Day', 'Month', 'AQI'])
  cols= inp_data.columns[1: ]
  len_cols= len(cols)
  op_data['Day']= inp_data['Date']
  op_data['Month']= month

  aqi_values = []
  for i in range(len(inp_data)):
    value= 0
    for j in range(len(cols)):
      value+= inp_data.iloc[i, j+1]
    aqi_values.append(int(value/len_cols))
    op_data['Day'] = op_data['Day'].astype(int)

  op_data['AQI']= aqi_values

  return op_data
```
